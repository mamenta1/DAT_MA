### 2.  Telling the story across majors (total popularity and popularity within women) in different ways. Please use the slides presented in class accordingly to select the story you'd want to display by creating the chart and placing an appropriate title.

#### 2.A We want to understand which majors are popular among women vs men.

```python
# Create a chart that shows  volume of majors by # of students ordered by rank for each total and women
## Hint, you may want two charts

popular_tot = earnings[['Major','Total','Major_category']].sort_values('Total',ascending=False).head(10)
popular_xx = earnings[['Major','Women','Major_category']].sort_values('Women',ascending=False).head(10)

popular_tot.Major = popular_tot.Major.str.capitalize()
popular_xx.Major = popular_xx.Major.str.capitalize()

ax = earnings.sort_values('Total',ascending=False).head(10).plot(x='Major', y='Total', kind='barh', figsize=(8,4));
ax.set_title('Most Popular Major', fontsize=14, y=1.01);
ax.legend(loc=1);
ax.set_ylabel('Students', fontsize=16);
ax.set_xlabel('Major', fontsize=16);

ax2 = popular_xx.plot(x='Major', kind='barh', figsize=(8,4));
ax2.set_title('Most Popular Major - Women', fontsize=14, y=1.01);
ax2.legend(loc=1);
ax2.set_ylabel('Women', fontsize=16);
ax2.set_xlabel('Major', fontsize=16);
```


#### 2.A.1 Challenge: We want to see clearly which major categories are impacted.

```python
# Iterate over 2.A.1. Choose colors of the bars based on major category
def color_mapper(row):
    if row['Major_category'] == 'Biology & Life Science':
        return 'blue'
    if row['Major_category'] == 'Business':
        return 'red'
    if row['Major_category'] == 'Communications & Journalism':
        return 'green'
    if row['Major_category'] == 'Health':
        return 'y'
    if row['Major_category'] == 'Humanities & Liberal Arts':
        return '0.5'
    if row['Major_category'] == 'Psychology & Social Work':
        return 'c'
    if row['Major_category'] == 'Social Science':
        return 'k'
    return 'w'

popular_tot.apply(lambda row: color_mapper(row), axis=1)

popular_tot['color'] = popular_tot.apply(lambda row: color_mapper(row), axis=1)

popular_tot

popular_tot.plot(x='Major', y='Total', kind='barh', figsize=(8,4), color = popular_tot.color);

#stumped
```

#### 2.A.2. We want to see the general picture

```python
# Create a chart that shows  volume of major categories # of students ordered by rank for each total and women
ax = popular_tot.plot(x='Major_category', kind='barh', figsize=(8,4));
ax2 = popular_xx.plot(x='Major_category', kind='barh', figsize=(8,4));
```

#### 2.B We are more interested in comparison of composition (instead of volume) total vs. women by major category. 

```python
# Create a chart that captures composition across major category, total vs. women
ax = earnings.sort_values('ShareWomen',ascending=False).head(10).plot(x='Major', y='ShareWomen', kind='barh', figsize=(8,4));
```


#### 2.C We want to understand if there is a correlation between popularity of total for all students vs. women.

```python
# Create a chart that determines relationship of popularity of major categories among men and women
## Hint, we would want multiple charts that share axes
earnings_cat = earnings.groupby('Major_category')['Total','Women'].sum()

earnings_cat.plot(x='Total', y='Women', kind='scatter', figsize=(10, 8));
```
